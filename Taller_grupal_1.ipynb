Definición función comparador

```python
#Usa dos cadenas y las compara caracter por caracter
def comp(a, b):
    for i in range(len(a)):
        if a[i] != b[i]: return False
    return True
```

Definición de la función de busqueda fuerza bruta

```python
#Cuenta la cantidad de coincidencias entre una cadena dada (txt) y una objetivo (trg)
def sbf(txt, trg):
    #tamaño de trg y cantidad de coincidencias
    len_trg = len(trg)
    matches = []

    #Ciclo que itera la posicion inicial de busqueda en la cadena txt
    for i in range(len(txt)-(len_trg-1)):
        #'fin' se usa para recortar la cadena txt al tamaño de trg, avanzando desde 0
        # hasta el tamaño de txt menos el tamaño de trg en cada ciclo.
        fin = i+len_trg

        if comp(txt[i:fin], trg):     #Busca si la cadena recortada es igual a trg
            matches.append((i,fin-1)) #De ser asi la agrega a 'matches' junto con el numero de iteracion del ciclo (Que corresponde a la posicion de la cadena trg en txt)
    
    if not matches:                   #Si no hay ninguna coincidencia retorna -1
        matches.append(-1)

    return matches
```

Ejemplo Bioinformático de busqueda por metodo de fuerza bruta (MAIN)


```python
#Se lee y guarda la secuencia a analizar
secuence=''
for l in lines[1:len(lines)-1]: #Se elimina la primera linea de enunciado
  secuence+=l[:len(l)-1]        #Se elimina el salto de linea, dejando una unica cadena "secuence"
print(secuence)
```
