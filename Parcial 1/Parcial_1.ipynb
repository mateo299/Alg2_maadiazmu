Punto 1:
Este problema plantea la solucion a una multiplicacion usando el metodo Rossen
Para esto, y  durante toda la ejecucion del algoritmo solo es necesario el uso de un solo ciclo, todo lo demas se podria despreciar al ser tiempos de ejecucion constantes.
El tiempo de ejecucion de este ciclo se define por el valor de una de las entradas, (n) y este valor se ve modificado en cada ciclo al ser dividido por 2, asi el tiempo es O(n/2) el cual es el mas significativo de todo el algoritmo


```python
import time

def multi(a, b): #Función que multiplica dos enteros positivos menores a 9999 por el metodo rossen
    if '.' in a or '.' in b: #Verificación para fload
        print('Numero no valido')
        return
    try:  #Verificación para no números
        a = int(a)
        b = int(b)
    except:
        print('Numero no valido')
        return
    if a > 99999 or b > 99999 or a < 0 or b < 0: #Verificación para positivo y menor o igual a 99999
        print('Numero no valido')
        return
    mi = min((a,b))
    ma = max((a,b))
    suma = 0
    while mi >= 1: #Metodo Rosse
        if mi % 2 == 1:
            suma += ma
        mi = int(mi/2)
        ma = ma*2
    return '{} x {} = {}'.format(a,b,suma) #Formateo de salida

def main():
    print('Ingrese dos numeros enteros a multiplicar positivos y menores o iguales a 99999 ')
    a = input('Ingrese primer numero ')
    b = input('Ingrese segundo numero ')

    #Tiempo de ejecucion
    tiempo=time.time()
    print(multi(a, b))
    #Tiempo de ejecucion
    print("Segundos desde inicio de ejecucion: ",time.time()-tiempo)


main()
```

**Punto 2**

Este problema plantea generar una cadena de texto usando dos "diccionarios" diferentes, uno con 10 carácteres y otro con 26. Al evaluar la complejidad de este algoritmo, podremos notar que la complejidad de la primera parte del problema depende del tamaño del primer diccionario (n) y del segundo (m), dando una complejidad de **O(n^3*m^3)**

```python
import time
def generateCrackedPassword(password):
    counter = 0
    crackedPassword = ''
    #Anidamos seis bucles para generar la contraseña ya que, aunque se podrían usar hilos, el problema no lo especifica.
    for i in range(0,9):
        for j in range(0,9):
            for k in range(0,9):
                for l in range(0, 26):
                    for m in range(0,26):
                        for n in range(0, 26):
                                counter =  counter + 1
                                #48 is the ASCII value for '0' and 97 the ASCII number for 'a'
                                crackedPassword = chr(i+48) + chr(j+48) + chr(k+48) + chr(l+97) + chr(m+97) + chr(n+97)
                                if crackedPassword == password:
                                    print("Cracked password in " + str(counter) + " iterations")
                                    return crackedPassword


def main():
    passwordToCrack = str(input("Enter password to crack: "))
    #Tiempo de ejecucion
    tiempo=time.time()
    generateCrackedPassword(passwordToCrack)
    #Tiempo de ejecucion
    print("Segundos desde inicio de ejecucion: ",time.time()-tiempo)

main()
```
